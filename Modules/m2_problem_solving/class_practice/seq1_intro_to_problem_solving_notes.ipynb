A number is called factor of N if it can divide N completely. so we naturally start from 1 and keep doing it till the number itself.

```python
def count_factors(n):
    factors = 0
    for i in range(1, n+1):
        if n%i == 0:
            factors += 1
    return factors
```

### General online compiler config 

CPU -> 10^8 iterations/sec <br>
Assuming 1 iteration is equal to 10 instructions.

N = 10^8 -> 10^8 iterations -> 10^8/10^8 => 1 sec

N = 10^9 -> 10^9 iterations -> 10^9/10^8 => 10 seconds

N = 10^18 -> 10^18 iterations -> 10^18/10^8 => 10^10 seconds => 317 years


How do we optimize??

```python
def count_factors(n):
    factors = 0
    for i in range(1, int(n**0.5)+1):
        x = i
        y = n/i
        if n%i == 0:
            if x == y:
                factors += 1
            else:
                factors += 2
    return factors
```

> Q. Check if a no. is a prime number?

Prime number: A number divisible by exactly only 2 numbers. first number is 1 and second number is itself.
Basically it should have 2 factors.

How many numbers are prime?
10 11 23 2 25 27 31

```python
def check_prime(n):
    factors = count_factors(n)
    if factors == 2:
        print("Prime number")
        return True
    else:
        print("Non Prime")
        return False
```

```python
check_prime(16)
check_prime(12)
check_prime(17)
```
