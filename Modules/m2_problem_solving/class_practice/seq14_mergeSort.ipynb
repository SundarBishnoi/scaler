```python
arr = [1, 3, 15, 6, 8, 9, 12, 18, 0]
# A, B = [1, 3, 15], [6, 8, 9, 12, 18]
def merge(A, B):
    i = 0
    j = 0
    k = 0
    # print(i, j, k)
    out = [0]*(len(A)+len(B))
    while i < len(A) and j < len(B):
        if A[i] <= B[j]:
            out[k] = A[i]
            i += 1
            k += 1
        elif B[j] < A[i]:
            out[k] = B[j]
            j += 1
            k += 1

    while i < len(A):
        out[k] = A[i]
        i += 1
        k += 1
    
    while j < len(B):
        out[k] = B[j]
        j += 1
        k += 1
    
    return out

def mergeSort(arr):
    if len(arr) == 1: return arr
    m = len(arr)//2
    A = arr[:m]
    B = arr[m:]
    A = mergeSort(A)
    B = mergeSort(B)
    out = merge(A, B)
    return out

# print(A, B)
# print(merge(A, B))

print(mergeSort(arr))

# TC: O(NlogN)
# SC : O(N + logN) => O(N)
```

> Stable sort - Respecting the insertion order if the magnitude is the same.

> Inplace sorting - No extra space is consumed.. thats why merge sort is not in place sort.

```python
# Javascript object based notation (web based applications)
import json
d = {"a": 1, "b":2}
# To convert dict to json string object
# Basically dumps will save and load is used to retrieve
json_out = json.dumps(d)
print(json_out)
print(type(json_out))
print(json_out[3:])


file_name = 'person.json'
# To convert dict to json object
# --- Writing to a JSON file using json.dump() ---
with open(file_name, 'w') as json_file:
    json.dump(d, json_file, indent=4)
    print(type(json_file))

# --- Reading from the JSON file using json.load() ---
# Open the file in read mode ('r')
with open(file_name, 'r') as json_file:
    loaded_data = json.load(json_file)

print(type(loaded_data))
print(loaded_data)
print(loaded_data['a'])
```

```python
# to write and read data from configuration.yaml
import yaml

data = {
    'name': 'Alice',
    'age': 30,
    'skills': ['Python', 'Data Analysis', 'Machine Learning'],
    'education': {
        'degree': 'Master\'s',
        'field': 'Computer Science',
        'year': 2021
    }
}

with open('data.yaml', 'w') as file:
    yaml.dump(data, file)
    print(file)
    print(type(file))


with open('data.yaml', 'r') as file:
    configurations = yaml.safe_load(file)
    print(configurations)
    print(type(configurations))
    print(configurations['name'])
    print(configurations['education']['field'])
```
