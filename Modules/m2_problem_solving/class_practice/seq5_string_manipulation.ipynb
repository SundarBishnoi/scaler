```python
s = "AcdZjKLPr"
def toggleCase(s):
    res = []
    for char in s:
        if ord(char) >= 65 and ord(char) <= 90:
            res.append(chr(ord(char)+32))
        elif ord(char) >= 97  and ord(char) <= 122:
            res.append(chr(ord(char)-32))

    print("".join(res))

toggleCase(s)
```

```python
def longestPS(s):
    ans = 0
    for c in range(len(s)):

        odd_ps = max_PS(s,c,c)

        even_ps = max_PS(s,c,c+1)

        ans = max(ans, odd_ps, even_ps)
    
    print(ans)


def max_PS(s, i , j):

    while(i>=0 and j < len(s)):
        if s[i] != s[j]: return j-i-1
        i -= 1
        j += 1
    
    return j-i-1

s = "anamadamspe"
s1 = "anamaddamspe"
longestPS(s)
longestPS(s1)
```

```python
# Find the maximum length palindromic substring around given center

s = "anamadamspe"
c = 5
n = len(s)
mx_ss = 0
i =c, j = c
while (i >= 0 and j < n):
    if s[i] == s[j]:
        mx_ss += 1
        i-=1
        j+=1
    else:
        break

print(mx_ss)


```
