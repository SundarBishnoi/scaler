```python
#find the frequency of elements provided in queries
arr = [2,6,3,8,2,8,2,3,8,10,6]
queries = [2,3,8,5]

def count_freq(arr, queries):
    fm = {}
    for ele in arr:
        fm[ele] = fm.get(ele,0)+1
    print(fm)

    for q in queries:
        print(fm.get(q,0))
count_freq(arr, queries)
```

```python
# Given a list of integers, find the number of unique elements...

arr = [1,2,1,3,2]

def countUnique(arr):
    # return len(set(arr))
    #  Or 
    s = set()
    for ele in arr:
        s.add(ele)
    return len(s)
countUnique(arr)
```

```python
# given a stream of integers, find the first non-repeating integers..

arr = [500,200,200,404,500,403]

fm = {}

for ele in arr:
    fm[ele] = fm.get(ele,0) + 1

print(fm)

for ele in arr:
    if fm[ele] == 1:
        print(ele)
        break
```

```python
# Given a list of integers, find the integers appearing more than N times..

arr = [1,5,1,2,1,5,5] 
N = 2
def getIntegersMoreThanNTimes(arr, N):
    fm = {}
    for ele in arr:
        fm[ele] = fm.get(ele,0)+1
    output = []
    for keys in fm:
        if fm[keys] > 2:
            output.append(keys)
    
    return output

getIntegersMoreThanNTimes(arr, N)
```
