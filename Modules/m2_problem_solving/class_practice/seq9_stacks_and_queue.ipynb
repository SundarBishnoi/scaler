```python
A = "))))))"
par = []
def check(A):
    for char in A:
        if char == "{" or char == "(" or char == "[" :
            par.append(char)
        if char == "}":
            if par and par[-1] == "{" :
                par.pop()
        if char == ")" :
            if par and par[-1] == "(":
                par.pop()
        if char == "]":
            if par and par[-1] == "[" :
                par.pop()

    if par:
        return 0
    else: return 1

print(check(A))
```
