```python
A = [1, 2, 3]
n = len(A)
lower_index_sum = []
higher_index_sum = [0]*n
for i in range(n):
    if i == 0:
        lower_index_sum.append(0)
    else:
        lower_index_sum.append(A[i-1] + lower_index_sum[i-1])
for i in reversed(range(n)):
    if i == (n-1):
        higher_index_sum[i] = 0
    else:
        higher_index_sum[i] = A[i+1] + higher_index_sum[i+1]

for i in range(n):
    if lower_index_sum[i] == higher_index_sum[i]:
        print(i)
    else:
        print(-1)
    
```
