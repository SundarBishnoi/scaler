Given an array of integers, find total sum of all possible subarrays


```python

arr = [1,2,3]

total = 0
n = len(arr)
for s in range(n):
    for e in range(s, n):
        for k in range(s, e+1):
            total += arr[k]
        
print(total)

# Brute force
#TC : O(N^3), SC: O(1)
```

```python
arr = [1,2,3]
n = len(arr)

prefx_sum = []
prefx_sum.append(arr[0])
for i in range(1,n):
    prefx_sum.append(prefx_sum[i-1]+arr[i])

print(prefx_sum)
total = 0
for s in range(n):
    for e in range(s, n):
        if s==0:
            total += prefx_sum[e]
        else:
            total += prefx_sum[e] - prefx_sum[s-1]
        

print(total)

#Using prefix_sum
#TC : O(N^2), SC: O(N)
```

```python
arr = [1,2,3]
n = len(arr)

ans = 0
for s in range(n):
    temp_sum = 0
    for e in range(s, n):
        temp_sum += arr[e]
        ans += temp_sum

print(ans)

# Using carry forward
#TC: O(N^2), SC: O(1)
```

To get the frequency of 1 in all subarray of lenght N array
[0,0]
[0,1]
[0,2]

(i+1)*(N-i)
(0+1)*(3-0) = 3

To get the sum of all subarray.. => num * frequency of num in all subarrays
 

```python
arr = [1,2,3]
n = len(arr)

ans = 0
for i in range(n):
    s = i+1
    e = n-i
    freq = s*e
    contri = arr[i]*freq
    ans += contri

print(ans)

# Contriubtion Techinique -> Google's favorite question
# TC: O(N), SC: O(1)
```

Given a value of K, how many subarray can we form of size K.

ar = [1, 2, 3, 4, 5, 6, 7]
K = 3


Brute Force <br>
Iterate and calculate sum for each window.



```python
ar = [1, 2, 3, 4, 5, 6, 7]
K = 3

s = 0
e = K-1
n = len(ar)
ans = float("-inf")

while e < n:
    w_sum = 0
    for i in range(s, e+1):
        w_sum += ar[i]
    
    s += 1
    e += 1
    
    ans = max(ans, w_sum)

print(ans)
```

```python
# Find sum of first K number
# Move the sliding window by 1(Add new element and suntract first element)

ar = [1, 2, 3, 4, 5, 6, 7]
K = 3


w_sum = 0
for i in range(K):
    w_sum += ar[i]

n = len(ar)

ans = w_sum
s=1
e=K

while e < n:
    w_sum = w_sum + ar[e]-ar[s-1]
    ans = max(ans, w_sum)
    s+=1
    e+=1

print(ans)
```
