```python
arr = [3, 6, 9, 12, 15, 18, 20, 23, 25, 27]
k = 12

def binarySearch(arr, k):
    l = 0 
    h = len(arr)-1
    ans = -1
    while l <= h:
        mid = (l+h)//2
        if arr[mid] < k:
            l = mid + 1
        elif arr[mid] > k:
            h = mid -1
        else:
            return mid
    
    return -1

binarySearch(arr, k)
```
