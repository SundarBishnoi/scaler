```python
def leftSearch(A, B):
    l = 0
    h = len(A)-1
    left_most = -1
    while l <= h:
        m = (l+h)//2
        if A[m] > B:
            h = m-1
        elif A[m] < B:
            l = m+1
        else:
            if m > 0 and A[m-1] == B:
                left_most = m-1
                h = m-1
            else:
                left_most = m
                break
    
    return left_most

def rightSearch(A, B):
    l = 0 
    h = len(A)-1
    right_most = -1
    while l <= h:
        m = (l+h)//2
        if A[m] > B:
            h = m-1
        elif A[m] < B:
            l = m+1
        else:
            if m < len(A)-1 and A[m+1] == B:
                right_most = m+1
                l = m+1
            else:
                right_most = m
                break
    
    return right_most
# A = [5, 7, 7, 8, 8, 10]
# B = 8
A = [ 1 ]
B = 1
left = leftSearch(A, B)
right = rightSearch(A, B)
out = []
out.append(left)
out.append(right)
print(out)
```
