Given a matrix of size of n*m, print all the diagnols from right to left(Anti-Diagnols)

```python
def healthcheck_status(log):
    for r in range(len(log)):
        for c in range(len(log[0])):
            if log[r][c] == 0:
                for i in range(len(log[0])):
                    if log[r][i] != 0:
                        log[r][i] = -1
                for i in range(len(log)):
                    if log[i][c] != 0:
                        log[i][c] = -1

    for r in range(len(log)):
        for c in range(len(log[0])):
            if log[r][c] == -1:
                log[r][c] = 0


    return log
```

Optimize by saving 0th row and column in a set and finally converting that row and col to zero..

```python
def healthcheck_status(log):
    zeroth_row = []
    zeroth_col = []
    for r in range(len(log)):
        for c in range(len(log[0])):
            if log[r][c] == 0:
                zeroth_row.append(r)
                zeroth_col.append(c)

    for row in zeroth_row:
        for i in range(len(log[0])):
            log[row][i] == 0

    for col in zeroth_col:
        for i in range(len(log)):
            log[i][col] = 0

    return log
```
