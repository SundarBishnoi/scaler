```python

def dec(A):
    if A == 0:
        return 
    print(A, end = " ")
    return dec(A-1)


def inc(A):
    if A == 0:
        return 
    inc(A-1)
    print(A, end = " ")
    

A = 4

dec(A)
inc(A)
print()
```
